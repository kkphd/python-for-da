## **JSON Data Extraction**

Using the legislators social media JSON file, https://theunitedstates.io/congress-legislators/legislators-social-media.json create a Pandas data frame which has the bioguideID, Twitter, Facebook, YouTube and Instagram handles/usernames for each of the members of Congress. 

Each of these should be stored as variables.








```
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup

url = 'https://www.realclearpolitics.com/epolls/other/president_trump_job_approval-6179.html'
html = urlopen(url).read()

soup = BeautifulSoup(html, 'lxml')
table = soup.find('table')

headings = [th.get_text() for th in table.find('tr').find_all('th')]

data_simple = []
for tr in table.find_all('tr'):
    row = []
    for td in tr:
        row.append(td.get_text())
    data_simple.append(row) 
data_simple = data_simple[1:]

data_df = pd.DataFrame(data_simple)
data_df.columns = headings

data_df['Poll'][0] = 'RCP Average'
data_df['Poll'][1] = 'Rasmussen'
data_df['Poll'][2] = 'YouGov'
data_df['Poll'][3] = 'Politico'
data_df['Poll'][4] = 'Reuters'
data_df['Poll'][5] = 'FOX News'
data_df['Poll'][6] = 'CNBC'
data_df['Poll'][7] = 'Quinnipiac'
data_df['Poll'][8] = 'Harris'

data_df
```

```
# JSON Data Extraction

import pandas as pd 
import numpy as np
import json
import requests

legsocial = requests.get('https://theunitedstates.io/congress-legislators/legislators-social-media.json').json()

legsocial_initial = []

for i in legsocial:
  try:
    bioguideID = i['id']['bioguide']
    twitterID = i['social']['twitter']
    fbID = i['social']['facebook']
    ytID = i['social']['facebook']
    instaID = i['social']['instagram']
    dict = {'Bioguide ID': bioguideID,
            'Twitter ID': twitterID,
            'Facebook ID': fbID,
            'YouTube ID': ytID,
            'Instagram ID': instaID }
    legsocial_initial.append(dict)
  except:
    next

legsocial_final = pd.DataFrame(legsocial_initial)
legsocial_final.head(3)
```
